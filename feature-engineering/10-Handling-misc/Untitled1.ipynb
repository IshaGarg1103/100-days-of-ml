```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
```

```python
url = 'https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day40-iterative-imputer/50_Startups.csv'
df = np.round(pd.read_csv(url)[['R&D Spend','Administration','Marketing Spend','Profit']]/10000)
np.random.seed(9)
df = df.sample(5)

df
```

```python
df = df.iloc[:,0:-1]
df
```

```python
df.iloc[1,0] = np.nan
df.iloc[3,1] = np.nan
df.iloc[-1,-1] = np.nan
```

```python
df.head()
```

```python
# Step 1 - Impute all missing values with mean of respective col

df0 = pd.DataFrame()
df0['R&D Spend'] = df['R&D Spend'].fillna(df['R&D Spend'].mean())
df0['Administration'] = df['Administration'].fillna(df['Administration'].mean())
df0['Marketing Spend'] = df['Marketing Spend'].fillna(df['Marketing Spend'].mean())
```

```python
# 0th iteration
df0
```

```python
#remove the coll imputed value
df1 = df0.copy()

df1.iloc[1,0] = np.nan

df1
```

```python
# use first 3 rows to build a model and use the last for prediction

x = df1.iloc[[0,2,3,4],1:3]
x
```

```python
y = df1.iloc[[0,2,3,4],0]
y
```

```python
df1.iloc[1,1:].values.reshape(1,2)
```

```python
lr = LinearRegression()
lr.fit(x,y)
lr.predict(df1.iloc[1,1:].values.reshape(1,2))
```

```python
df2.iloc[3,1] = np.nan
X = df2.iloc[[0,1,2,4],[0,2]]
y = df2.iloc[[0,1,2,4],1]
lr = LinearRegression()
lr.fit(X,y)
m = lr.predict(df2.iloc[3,[0,2]].values.reshape(1,2))
df2.iloc[3,1] = m
df3 = df2.copy()
    
df3.iloc[1,0] = np.nan
X = df3.iloc[[0,2,3,4],1:3]
y = df3.iloc[[0,2,3,4],0]
lr = LinearRegression()
lr.fit(X,y)
m = lr.predict(df3.iloc[1,1:].values.reshape(1,2)
df3.iloc[1,0] = m
    
df3.iloc[3,1] = np.nan
X = df3.iloc[[0,1,2,4],[0,2]]
y = df3.iloc[[0,1,2,4],1]
lr = LinearRegression()
lr.fit(X,y)
m = lr.predict(df3.iloc[3,[0,2]].values.reshape(1,2))
df3.iloc[3,1] = m

df3.iloc[4,-1] = np.nan
X = df3.iloc[0:4,0:2]
y = df3.iloc[0:4,-1]
lr = LinearRegression()
lr.fit(X,y)
m = lr.predict(df3.iloc[4,0:2].values.reshape(1,2))
df3.iloc[4,-1] = m
```
