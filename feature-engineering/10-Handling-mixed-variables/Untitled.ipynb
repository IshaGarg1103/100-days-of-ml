```python
import numpy as np
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day33-handling-mixed-variables/titanic.csv'
df = pd.read_csv(url)
```

```python
# extract numerical part
df['number_numerical'] = pd.to_numeric(df['number'],errors = 'coerce',downcast = 'integer')
```

error = 'coerce'
If you can’t convert a value to a number, replace it with NaN instead of raising an error.

downcast='integer'
After converting to numeric, try to use the smallest possible integer type (like int8 or int16) to save memory.

```python
#extract categorical part
df['number_categorical'] = np.where(df['number_numerical'].isnull(),df['number'],np.nan)
```

df['number_numerical'].isnull()

- Returns True where 'number_numerical' is NaN 

- Returns False where 'number_numerical' has valid numbers.

np.where(df['number_numerical'].isnull(), df['number'], np.nan)

- If number_numerical is NaN → take the original value from df['number']

- Else → assign NaN

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)')
df['cabin_cat'] = df['Cabin'].str[0] #captures the first letter
```

```python
df.head()
```

- ()	defines what to capture and return
- \d+	means “one or more digits”
- (\d+)	means “capture one or more digits so I can extract them”

```python
#Extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                 errors = 'coerce',
                                 downcast = 'integer')
```

df['Ticket'] → accesses the Ticket column (which contains strings like "PC 17599" or "A/5 21171").

- .apply(lambda s: ...) → applies a small anonymous function (lambda) to each ticket string.

Inside the lambda:

- s.split() → splits the string by spaces into a list.

- - Example: "A/5 21171" → ['A/5', '21171']

- [-1] → takes the last element of that list.

- - So you get '21171' from ['A/5', '21171'].

```python
#extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s : s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(),np.nan,df['ticket_cat'])
```
