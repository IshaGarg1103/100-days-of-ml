```python
import numpy as np
import pandas as pd

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day45-feature-construction-and-feature-splitting/train.csv'
df = pd.read_csv(url)[['Age','Pclass','SibSp','Parch','Survived']]
```

```python
df.head()
```

```python
df.dropna(inplace = True )
#This removes the rows with NaN from the original df itself.
```

```python
df.head()
```

```python
x = df.iloc[:,0:4]
y = df.iloc[:,-1]
```

```python
x.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),x,y,scoring = 'accuracy',cv = 20))
```

## Applying Feature Construction

```python
x['Family_size'] = x['SibSp'] + x['Parch'] + 1
#1 - is for the passenger itself
```

```python
x.head()
```

```python
def myfunc(num) :
    if num == 1:
        #alone
        return 0
    elif num > 1 and num <= 4 :
        #small family
        return 1
    else:
        #large family
        return 2
    
```

```python
myfunc(4)
```

```python
x['Family_type'] = x['Family_size'].apply(myfunc)
#- takes each value from the column Family_size
#- pass it to myfunc
#- store the returned label in a new column Family_type
```

```python
x.head()
```

```python
x.drop(columns = ['SibSp','Parch','Family_size'],inplace = True)
```

```python
x.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),x,y,scoring = 'accuracy',cv = 20))
```

## Feature Splitting

```python
url = 'https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day45-feature-construction-and-feature-splitting/train.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df['Name']
```

```python
df['Name'].str.split(', ',expand = True)
```

```python
df['Title'] = df['Name'].str.split(', ',expand = True)[1].str.split('.',expand = True)[0]
#(expand = true)-This turns the split result into separate DataFrame columns
```

```python
df[['Title','Name']]
```

```python
df.groupby('Title')['Survived'].mean().sort_values(ascending = False)
```

```python
df['Is_married'] = 0
df['Is_married'].loc[df['Title'] == 'Mrs'] = 1
```
