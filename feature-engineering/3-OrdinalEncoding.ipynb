```python
import numpy as np
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day26-ordinal-encoding/customer.csv'
df = pd.read_csv(url)
```

```python
df.sample(5)
```

gender is nominal = one hot encoder
review and education is ordinal = ordinal encoder
purchased is nominal categorical = label encoder

```python
df = df.iloc[:,2 :]
```

```python
df.head(15)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size = 0.2)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = [['Poor','Average','Good'],['School','UG','PG']]) #while transformation give values according to the order.
```

```python
oe.fit(x_train)
```

```python
x_train = oe.transform(x_train)
x_test  = oe.transform(x_test)
```
