```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:,0:4], df.iloc[:,-1],test_size = 0.2,random_state = True)
```

## 4. OneHotEncoding with Top Categories

```python
counts = df['brand'].value_counts()
```

```python
df['brand'].nunique()
threshold = 100
```

```python
counts[counts<= threshold].index
```

```python
repl = counts[counts<= threshold].index
```

```python
df['brand'].replace(repl,'uncommon')
```

```python
pd.get_dummies(df['brand'].replace(repl, 'uncommon'),dtype = int).sample(5)
```
