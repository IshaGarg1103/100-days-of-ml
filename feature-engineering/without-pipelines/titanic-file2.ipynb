```python
import pickle
import numpy as np
import pandas as pd
```

```python
#decode
ohe_sex = pickle.load(open('models/ohe_sex.pkl','rb'))
ohe_embarked = pickle.load(open('models/ohe_embarked.pkl','rb'))
clf = pickle.load(open('models/clf.pkl','rb'))
```

```python
#assume user input
#Pclass/gender/age/SibSp/fare/embarked
test_input= np.array([2,'male',31.0,0,0,10.5,'S'],dtype = object).reshape(1,7)
#after reshape(1,7)
#it becomes from [2, 'male', 31.0, 0, 0, 10.5, 'S'] -> [[2, 'male', 31.0, 0, 0, 10.5, 'S']]
```

```python
test_input
```

```python
cols = ['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']
test_input_df = pd.DataFrame([test_input.flatten()],columns = cols)
```

```python
test_input_sex = ohe_sex.transform(test_input_df[['Sex']])
test_input_sex
```

```python
test_input_embarked = ohe_embarked.transform(np.asarray(test_input_df[['Embarked']]))
test_input_embarked
```

```python
test_input_age = test_input[:,2].reshape(1,1)
test_input_age
```

```python
test_input_transformed = np.concatenate((test_input[:,[0,3,4,5]],test_input_age,test_input_sex,test_input_embarked),axis=1)
```

```python
test_input_transformed.shape
```

```python
clf.predict(test_input_transformed)
```
