```python
response=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8156ba7859bd0552330c4491d2d52d0c&language=en-US&page=1')
```

```python
df=pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
```

```python
df=pd.DataFrame()
type(df)
```

```python
for i in range(1,501):
    response=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8156ba7859bd0552330c4491d2d52d0c&language=en-US&page={}'.format(i))
    temp_df=pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df=pd.concat([df,temp_df],ignore_index=True)
#when you run the loop runs first time, it would give index from 1-20, but for second itertaion it will again give 1-20, so to ignore that, it will giving indexing continuing from 21 onwards.
```
