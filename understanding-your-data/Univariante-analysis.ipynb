```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

## 1. Categorical Data

## a. Countplot

```python
sns.countplot(x='Pclass', hue='Pclass', data=df,palette="Set2", legend=False)
plt.show()
```

## b. PieChart

```python
df['Pclass'].value_counts().plot(kind='pie',autopct = '%.2f')
plt.ylabel('Passenger class')
plt.show()
```

## 2. Numerical Data

## a. Histogram

```python
plt.hist(df['Age'],bins = 10) #to make the bars range less or more
```

## b. Displot

```python
sns.displot(df['Age'],kde = 'True')
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Distribution of Age')
plt.show()
```

## c. Boxplot

```python
sns.boxplot(df['Age']) #when your data has noise.
```
