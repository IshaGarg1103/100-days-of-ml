403 code means that your request has been denied
APIs may block unknown/bot requests, so developers often add a User-Agent string to look like a real browser.

```python
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/139.0.0.0 Safari/537.36 Edg'}
webpage=requests.get('https://internshala.com/internships/?page=1',headers=headers).text
```

```python
soup = BeautifulSoup(webpage,'lxml') #to parse HTML
```

```python
print(soup.prettify())
```

```python
soup.find_all('h1')[0].text
```

```python
for i in soup.find_all('h3'):
    print(i.text.strip())#to remove the special character, like space and slash
```

```python
len(soup.find_all('p',class_= 'company-name'))
```

```python
len(soup.find_all('div', class_= 'actively-hiring-badge'))
```

here when we try to fetch if the 'company is hiring or not', not all the companies are provided with that batch, hence out of 50 companies, only 23 have that batch.

```python
company = soup.find_all('div', class_='internship_meta duration_meta')
```

hence, we now fetch the data of each container, directly, instead of individually extracting it.

```python
company
```

```python
company_name = []
job=[]
for i in company :
    job.append(i.find('h3').text.strip())
    company_name.append(i.find('p').text.strip())
job
```
